```python
import numpy as np
import pandas as pd

data = np.random.randn(4, 4)  # Generate random values
df = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'])
df
```

```python
new_column_names = {
    'Feature 1': 'Random Value 1',
    'Feature 2': 'Random Value 2',
    'Feature 3': 'Random Value 3',
    'Feature 4': 'Random Value 4'
}

df.rename(columns=new_column_names, inplace=True)
df
```

```python
statistics = df.describe()
print(statistics)
```

```python
null_values = df.isnull().sum()
data_types = df.dtypes

print("Null Values:")
print(null_values)
print("\nData Types:")
print(data_types)
```

```python
columns_loc = df.loc[:, ['Random Value 2', 'Random Value 3']]
print("Columns using location method (loc):\n", columns_loc)

columns_iloc = df.iloc[:, [1, 2]]
print("\nColumns using index location method (iloc):\n", columns_iloc)
```
